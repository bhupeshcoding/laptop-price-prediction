```python
y=data["Price"]
x=data[["Processor_Speed", "RAM_Size","Storage_Capacity"]]
```

```python
# Random Forest Regression is a machine learning algorithm used for predicting continuous numerical values.
```

```python
# The train_test_split() method is used to split our data into train and test sets
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
len(x_test)
```

```python
len(y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel=RandomForestRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Hyperparameter Tuning
param_gridrfr = {
    "max_depth": [2, 5, 10, 15, 20, 25],
    "max_features": ["auto", "log2"],
    "n_estimators": [2, 4, 6, 8, 10, 12]
}
```

```python
gridrrfr=GridSearchCV(rfrmodel,param_gridrfr)
```

```python
gridrrfr.fit(x_train, y_train)
```

```python
gridrrfr.best_params_
```

```python
predictions=gridrrfr.predict(x_test)
# predict method is used to predict the output for new input data.
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,predictions)
```
